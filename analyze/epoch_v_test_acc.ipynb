```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
name = "gsm8k_fft_easy50_constantlr"

checkpoints = [156, 312, 468, 624, 780]
easy50_accs = []

for checkpoint in checkpoints:
   answer_types_all = np.load(f"../ckpts/{name}/checkpoint-{str(checkpoint)}/test_answer_types.npy")
   acc = (answer_types_all==0).mean(axis=0)
#    print(acc)
   easy50_accs.append(acc)
   
easy50_accs = np.array(easy50_accs)
```

```python
easy50_accs
```

```python
name = "gsm8k_fft_hard50_constantlr"

checkpoints = [156, 312, 468, 624, 780]
hard50_accs = []

for checkpoint in checkpoints:
   answer_types_all = np.load(f"../ckpts/{name}/checkpoint-{str(checkpoint)}/test_answer_types.npy")
   acc = (answer_types_all==0).mean(axis=0)
#    print(acc)
   hard50_accs.append(acc)
   
hard50_accs = np.array(hard50_accs)
```

```python
hard50_accs
```

```python
name = "gsm8k_fft_rand50_constantlr"

checkpoints = [156, 312, 468, 624, 780]
rand50_accs = []

for checkpoint in checkpoints:
   answer_types_all = np.load(f"../ckpts/{name}/checkpoint-{str(checkpoint)}/test_answer_types.npy")
   acc = (answer_types_all==0).mean(axis=0)
#    print(acc)
   rand50_accs.append(acc)

rand50_accs = np.array(rand50_accs)
```

```python
rand50_accs
```

```python
name = "gsm8k_fft_full_constantlr"

checkpoints2 = [311, 623, 934, 1246, 1557, 1869, 2180, 2492, 2803, 3110]
full_accs = []

for checkpoint in checkpoints2:
   answer_types_all = np.load(f"../ckpts/{name}/checkpoint-{str(checkpoint)}/test_answer_types.npy")
   acc = (answer_types_all==0).mean(axis=0)
#    print(acc)
   full_accs.append(acc)
   
full_accs = np.array(full_accs)
```

```python
plt.scatter(np.tile([checkpoints], [10,1]).transpose(), (easy50_accs), alpha=0.5)
plt.plot(checkpoints, easy50_accs.mean(axis=-1), "o-", label = "easy50", linewidth=3, markersize=12, alpha=0.9)
plt.scatter(np.tile([checkpoints], [10,1]).transpose(), (hard50_accs), alpha=0.5)
plt.plot(checkpoints, hard50_accs.mean(axis=-1), "o-", label = "hard50", linewidth=3, markersize=12, alpha=0.9)
plt.scatter(np.tile([checkpoints], [10,1]).transpose(), (rand50_accs), alpha=0.5)
plt.plot(checkpoints, rand50_accs.mean(axis=-1), "o-", label = "rand50", linewidth=3, markersize=12, alpha=0.9)
plt.scatter(np.tile([checkpoints2], [10,1]).transpose(), (full_accs), alpha=0.5)
plt.plot(checkpoints2, full_accs.mean(axis=-1), "o-", label = "full", linewidth=3, markersize=12, alpha=0.9)
plt.legend()


plt.xlabel("checkpoint")
plt.ylabel("test acc")
plt.title("constantlr 5e-5")
```

```python
name = "gsm8k_fft_full_constantlr"

checkpoints2 = [311, 934, 1557, 3110]
full_accs = []

for checkpoint in checkpoints2:
   answer_types_all = np.load(f"../ckpts/{name}/checkpoint-{str(checkpoint)}/train_answer_types.npy")
   acc = (answer_types_all==0).mean(axis=0)
#    print(acc)
   full_accs.append(acc)
   
full_accs = np.array(full_accs)
```
