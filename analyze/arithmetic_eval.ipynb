```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
accs_all_hard_skip = []
for ckpt in [500, 5000, 10000, 20000, 30000]:
    accs_ckpt = []
    for i in range(2, 8):
        accs_ckpt.append((np.load(f"../ckpts/arit_fft_hard_skip{ckpt}/{i}step_eval_answer_types5.npy")==0).mean())
    accs_all_hard_skip.append(accs_ckpt)
accs_all_hard_skip = np.array(accs_all_hard_skip)
```

```python
accs_all_mixed = []
for ckpt in [500, 5000, 10000, 20000, 30000]:
    accs_ckpt = []
    for i in range(2, 8):
        accs_ckpt.append((np.load(f"../ckpts/arit_fft_mixed{ckpt}/{i}step_eval_answer_types5.npy")==0).mean())
    accs_all_mixed.append(accs_ckpt)
    
accs_all_mixed = np.array(accs_all_mixed)
```

```python
accs_all_easy_full = []
for ckpt in [500, 5000, 10000, 20000, 30000]:
    accs_ckpt = []
    for i in range(2, 8):
        accs_ckpt.append((np.load(f"../ckpts/arit_fft_easy_full{ckpt}/{i}step_eval_answer_types5.npy")==0).mean())
    accs_all_easy_full.append(accs_ckpt)
    
accs_all_easy_full = np.array(accs_all_easy_full)
```

```python
plt.plot([500, 5000, 10000, 20000, 30000], accs_all_hard_skip.mean(axis=-1), "o-", label = "hard skip")
plt.plot([500, 5000, 10000, 20000, 30000], accs_all_mixed.mean(axis=-1), "o-", label = "mixed")
plt.plot([500, 5000, 10000, 20000, 30000], accs_all_easy_full.mean(axis=-1), "o-", label = "easy full")

plt.legend()
plt.xlabel("Num train points")
plt.ylabel("Test acc")
plt.title("Synthetic arithmetic")
```

```python
f, ax = plt.subplots(2, 3, sharey=True, figsize=(10, 6))

for i in range(6):
    ax[i//3][i%3].plot([500, 5000, 10000, 20000, 30000], accs_all_hard_skip[:, i], "o-", label = "hard skip")
    ax[i//3][i%3].plot([500, 5000, 10000, 20000, 30000], accs_all_mixed[:, i], "o-", label = "mixed")
    ax[i//3][i%3].plot([500, 5000, 10000, 20000, 30000], accs_all_easy_full[:, i], "o-", label = "easy full")
    ax[i//3][i%3].set_title("Num operations: "+str(i+1))
    ax[i//3][i%3].set_ylabel("Test acc")
    ax[i//3][i%3].set_xlabel("Num train points")
```
